```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt_image = plt.imread('ASHE-PF.jpg')
plt.imshow(plt_image)
```

```python
def show_image(im, h=15, **kwargs):
    
    '''
    helper function to show image
    '''
    
    y = im.shape[0]
    x = im.shape[1]
    
    w = (y/x) * h
    
    plt.figure(figsize = (w,h))
    plt.imshow(im, interpolation = "none", **kwargs)
    plt.axis('off')
    
show_image(plt_image)
    
```

### Get rgb

```python
fig, axs = plt.subplots(nrows = 3, ncols = 1, figsize = (10, 20))

for c, ax in zip(range(3), axs): # r = 0, g = 1, b = 2
    tmp_im = np.zeros(plt_image.shape, dtype='uint8')
    tmp_im[:,:,c] = plt_image[:, :, c]
    ax.imshow(tmp_im)
    ax.set_axis_off()
```

### Color transformations

```python
def normalize_image(plt_image):
    return -np.log(1 / ((1 + plt_image) / 257) - 1)

def undo_normalize(plt_image):
    return (1 + 1 / (np.exp(-plt_image) + 1) * 257)

def rotation_matrix(theta):
    
    '''
    3D rotation matrix around the x axis by angle theta
    '''
    return np.c_[
        [1,0,0],
        [0, np.cos(theta), -np.sin(theta)],
        [0, np.sin(theta), np.cos(theta)]
    ]

im_normed = normalize_image(plt_image)
plt.imshow(im_normed)
```

```python
im_rotated = np.einsum("ijk,lk->ijl", im_normed, rotation_matrix(np.pi))
plt.imshow(im_rotated)
```

```python
im_undo_norm = undo_normalize(im_rotated)
show_image(im_undo_norm)
```

### Creating a colour rotation gif

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(10,18))

def update(i):
    im_normed = normalize_image(plt_image)
    imrotated = np.einsum("ijk, lk->ijl", im_normed, rotation_matrix(i * np.pi/10))
    im2 = undo_normalize(im_rotated)
    
    ax.imshow(im2)
    ax.set_titlele("Angle: {}*pi/10".format(i), fontsize = 20)
    ax.set_axis_off()
    
anim = FuncAnimation(fig, update, frames = np.arange(0,10), interval = 50)
anim.save('color_rotation.gif', dpi = 80, writer = 'imagemagick')
plt.close()
```

### To Grayscale

```python
def to_grayscale(im, weights = np.c_[0.2989, 0.5870, 0.1140]):
    '''
    transforms a color image to grayscale by taking mean of rgb values weighted by
    the matrix weights
    '''
    title = np.tile(weights, reps = (im.shape[0], im.shape[1], 1))
    return np.sum(title * im, axis = 2)
```

```python
img = to_grayscale(plt_image)

show_image(img, cmap='Greys')
```
